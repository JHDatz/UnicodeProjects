```python
import pandas as pd 

url = 'https://raw.githubusercontent.com/JHDatz/UnicodeProjects/main/Twitter%20Project/Data/clean.csv'
df = pd.read_csv(url)
```

```python
import numpy as np
import os
from google.colab import files
import matplotlib.pyplot as plt
import sklearn
import numpy as np

#others
from sklearn.metrics import f1_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix

from scipy.special import softmax
```

```python
df['clean'].sample(10).tolist()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace = True)
```

```python
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df['tweet'], df['injury_report'], random_state = 91, stratify = df['injury_report'], test_size = 0.2)
train = pd.concat([X_train, y_train], axis = 1)
test = pd.concat([X_test, y_test], axis = 1)
y = test['injury_report'].tolist()
```

```python
# import pandas as pd 

# url = 'https://raw.githubusercontent.com/Jhagrut/Twitter-Project/main/Data/train.csv'
# train = pd.read_csv(url)
# url = 'https://raw.githubusercontent.com/Jhagrut/Twitter-Project/main/Data/test.csv'
# test = pd.read_csv(url)

# Dataset is now stored in a Pandas Dataframe

train = train[['tweet','injury_report']]
test = test[['tweet','injury_report']]
tst = test['tweet'].tolist()
```

```python
train.reset_index().to_csv("train 25-08.csv")
files.download('train 25-08.csv')
test.reset_index().to_csv("test 25-08.csv")
files.download('test 25-08.csv')
```

```python
from simpletransformers.classification import ClassificationModel
import collections
```

```python
print(collections.Counter(train["injury_report"]))
print(collections.Counter(test["injury_report"]))
```

```python
bert = ClassificationModel('distilbert', 'distilbert-base-uncased', num_labels=2, use_cuda=True, args={
    'reprocess_input_data': True,
    'use_cached_eval_features': False,
    'overwrite_output_dir': True,
    'num_train_epochs': 5,
    'use_early_stopping': True,
    'early_stopping_delta': 0.01,
    'early_stopping_metric': "mcc",
    'early_stopping_metric_minimize': False,
    'early_stopping_patience': 5,
    'evaluate_during_training_steps': 1000
})
```

```python
bert
```

```python
bert.train_model(train)
```

```python
def evaluations(model, model_name):
  result, model_outputs, wrong_predictions = model.eval_model(test, acc=sklearn.metrics.accuracy_score)
  print("Result: ")
  print(result, end = "\n\n")
  
  df_model_outputs = pd.DataFrame(model_outputs, columns = ['P0','P1'])
  df_model_outputs.to_csv(model_name +'_model_outputs.csv')
  files.download(model_name +'_model_outputs.csv')
  print(model_name + 'outputs downloaded', end = '\n\n')

  print(wrong_predictions, end = '\n\n')
  
  fn = []
  fp = []
  for i in range(len(model_outputs)):
    x = model_outputs[i];
    if x[1]>x[0] and y[i]==0:
      fp.append([i,tst[i]])
    elif x[0]>x[1] and y[i]==1:
      fn.append([i,tst[i]])
  print(fp)
  print(len(fp))
  print(fn)
  print(len(fn))
  
  df_fp = pd.DataFrame(fp, columns = ['index','fp tweet'])
  df_fp.to_csv(model_name + '_fp_tweets.csv')
  files.download(model_name + '_fp_tweets.csv')

  df_fn = pd.DataFrame(fp, columns = ['index','fn tweet'])
  df_fn.to_csv(model_name + '_fn_tweets.csv')
  files.download(model_name + '_fn_tweets.csv')

  print(model_name + ' FP and FN tweet dataframes downloaded', end = '\n\n')
```

```python
result, model_outputs, wrong_predictions = bert.eval_model(test, acc=sklearn.metrics.accuracy_score)
print(result)
```
